### Plot the housing price

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Sample dataset
house_data = pd.read_csv('./house_data.csv')

# Convert to DataFrame
df = pd.DataFrame(house_data)

# Fit linear regression (Size vs Price)
slope, intercept = np.polyfit(df['Size'], df['Price'], 1)
regression_line = slope * np.array(df['Size']) + intercept

# Plot setup
plt.figure(figsize=(8, 5))
plt.scatter(df['Size'], df['Price'], color='red', label='Data Points')

# Plot regression line
plt.plot(df['Size'], regression_line, color='royalblue', linewidth=2, label='Regression Line')

# ----  Labels and Title  ----
# Format y-axis ticks as '100k', '200k', etc.

def thousands_formatter(x, pos):
    return f'{int(x/1000)}'
# x : is the tick value being formatted (like 700000, 400000, etc.).
# pos : is the tick position index along the axis.
# pos=0 for the first tick
# pos=1 for the second tick
# Even if you donâ€™t use pos inside the function, matplotlib always passes it because the formatter function needs to accept two arguments by design.


plt.gca().yaxis.set_major_formatter(FuncFormatter(thousands_formatter))

plt.xlabel('Size (sq ft)')
plt.ylabel('Price (1000 USD)')
plt.title('Housing Prices')

# Legend
plt.legend()

# Modern-minimal style tweaks
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()

# Save as vector file (SVG or PDF are good choices)
plt.savefig('house_price_plot.svg')    # or .pdf / .eps

# Show plot
plt.show()
```
